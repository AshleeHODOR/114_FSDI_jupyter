```python
# Pass by value vs pass by reference 

# In python3 there are 3 data types that pass by value: int, bool, float

x = 5
y = x
x += 1
print(id(x))
print(id(y))

# Other data types pass by reference

x = [1,2,3]
y = x
x.append(4)
print(id(x))
print(id(y))
#pass by refernce
# x/y SAME LIST

```
